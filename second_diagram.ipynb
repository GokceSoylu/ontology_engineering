```python
# Enhanced ontology visualization based on the extended analysis document

# Define classes (nodes) with updated labels for better readability
import networkx as nx
import matplotlib.pyplot as plt
classes_extended = [
    "Event", "Location", "Time", "Participant", "Organizer", 
    "Event Type", "City", "Venue", "Role", "Organization\nType", 
    "Start Date", "End Date"
]

# Define relationships (edges) based on competency questions and extended ontology
relationships_extended = [
    ("Event", "Location"),  # Event has Location
    ("Event", "Time"),      # Event has Time
    ("Event", "Participant"),  # Event has Participant
    ("Event", "Organizer"),  # Event has Organizer
    ("Event", "Event Type"),  # Event has Type
    ("Event", "Start Date"),  # Event has Start Date
    ("Event", "End Date"),    # Event has End Date
    ("Location", "City"),  # Location has City
    ("Location", "Venue"),  # Location has Venue
    ("Participant", "Role"),  # Participant has Role
    ("Organizer", "Organization\nType"),  # Organizer has Organization Type
    ("Participant", "Event"),  # Participant attends Event
    ("Organizer", "Event")  # Organizer creates Event
]

# Create a new directed graph
G_extended = nx.DiGraph()
G_extended.add_nodes_from(classes_extended)
G_extended.add_edges_from(relationships_extended)

# Set positions for nodes to create a clear and organized layout
pos_extended = {
    "Event": (0, 0),
    "Location": (-2, -2),
    "Time": (2, -2),
    "Participant": (-3, 0),
    "Organizer": (3, 0),
    "Event Type": (-1, 1),
    "City": (-3, -3),
    "Venue": (-1, -3),
    "Role": (-4, 0),
    "Organization\nType": (4, -1),
    "Start Date": (1, -1),
    "End Date": (3, -3)
}

# Draw the graph with extended features
plt.figure(figsize=(12, 10))
nx.draw(
    G_extended, pos_extended, with_labels=True,
    labels={node: node for node in G_extended.nodes()},  # Use node names as labels
    node_size=3500, node_color='lightblue', 
    font_size=10, font_weight='bold', arrows=True, edge_color='darkblue'
)

# Title and show
plt.title("Extended Event Ontology Diagram", fontsize=16)
plt.show()
```

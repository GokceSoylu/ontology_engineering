```python
# Enhanced ontology diagram based on the problem definition document
# Adding attributes to the existing classes and defining additional relationships
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G_advanced = nx.DiGraph()

# Add classes (nodes) to the graph with attributes
classes_advanced = [
    "Event", "Location", "Time", "Participant", "Organizer", 
    "Event Type", "City", "Venue", "Role", "Organization Type"
]
G_advanced.add_nodes_from(classes_advanced)

# Add relationships (edges) between the classes
relationships_advanced = [
    ("Event", "Location"),  # Event has Location
    ("Event", "Time"),      # Event has Time
    ("Event", "Participant"),  # Event has Participant
    ("Event", "Organizer"),  # Event has Organizer
    ("Event", "Event Type"),  # Event has Type
    ("Location", "City"),  # Location has City
    ("Location", "Venue"),  # Location has Venue
    ("Participant", "Role"),  # Participant has Role
    ("Organizer", "Organization Type"),  # Organizer has Organization Type
    ("Participant", "Event"),  # Participant attends Event
    ("Organizer", "Event")  # Organizer creates Event
]

G_advanced.add_edges_from(relationships_advanced)

# Set the position for each node for a clear visualization
pos_advanced = {
    "Event": (0, 0), "Location": (-2, -2), "Time": (2, -2), 
    "Participant": (-3, 0), "Organizer": (3, 0), "Event Type": (-2, 1), 
    "City": (-3, -3), "Venue": (-1, -3), "Role": (-4, 0), "Organization Type": (4, 0)
}

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(G_advanced, pos_advanced, with_labels=True, node_size=3000, node_color='lightgreen', 
        font_size=9, font_weight='bold', arrows=True, edge_color='blue')

# Title and show
plt.title("Advanced Event Ontology Diagram", fontsize=15)
plt.show()
```

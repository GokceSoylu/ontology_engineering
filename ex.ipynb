```python
# Daha karmak bir ontoloji yaps iin rnek veri oluturma
import networkx as nx
import matplotlib.pyplot as plt
complex_nodes = {
    "AI Conference": (0, 0), "Blockchain Summit": (-4, 2), "Music Festival": (4, 2),
    "Tech Center": (-2, -2), "Concert Hall": (2, -2), "Innovation Hub": (-6, 0),
    "New York": (-3, -3), "Los Angeles": (3, -3), "San Francisco": (-7, -1),
    "John Doe": (-4, 0), "Jane Smith": (-2, 1), "Alice Brown": (4, 0),
    "Tech Events Inc.": (-6, -2), "Music Mania Org.": (6, -2),
    "2024-12-01 to 2024-12-03": (0, -1), "2025-01-10 to 2025-01-12": (-4, -1),
    "2025-02-15 to 2025-02-16": (4, -1)
}

complex_edges = [
    # Events and Locations
    ("AI Conference", "Tech Center"), ("Blockchain Summit", "Innovation Hub"), ("Music Festival", "Concert Hall"),
    # Locations and Cities
    ("Tech Center", "New York"), ("Concert Hall", "Los Angeles"), ("Innovation Hub", "San Francisco"),
    # Events and Participants
    ("AI Conference", "John Doe"), ("AI Conference", "Jane Smith"), ("Music Festival", "Alice Brown"),
    ("Blockchain Summit", "John Doe"), ("Blockchain Summit", "Alice Brown"),
    # Events and Organizers
    ("AI Conference", "Tech Events Inc."), ("Blockchain Summit", "Tech Events Inc."),
    ("Music Festival", "Music Mania Org."),
    # Events and Times
    ("AI Conference", "2024-12-01 to 2024-12-03"),
    ("Blockchain Summit", "2025-01-10 to 2025-01-12"),
    ("Music Festival", "2025-02-15 to 2025-02-16")
]

# Create a directed graph for the complex ontology
G_complex = nx.DiGraph()
G_complex.add_nodes_from(complex_nodes.keys())
G_complex.add_edges_from(complex_edges)

# Relationships for edge labels
complex_edge_labels = {
    ("AI Conference", "Tech Center"): "hasLocation",
    ("Blockchain Summit", "Innovation Hub"): "hasLocation",
    ("Music Festival", "Concert Hall"): "hasLocation",
    ("Tech Center", "New York"): "city",
    ("Concert Hall", "Los Angeles"): "city",
    ("Innovation Hub", "San Francisco"): "city",
    ("AI Conference", "John Doe"): "hasParticipant",
    ("AI Conference", "Jane Smith"): "hasParticipant",
    ("Music Festival", "Alice Brown"): "hasParticipant",
    ("Blockchain Summit", "John Doe"): "hasParticipant",
    ("Blockchain Summit", "Alice Brown"): "hasParticipant",
    ("AI Conference", "Tech Events Inc."): "hasOrganizer",
    ("Blockchain Summit", "Tech Events Inc."): "hasOrganizer",
    ("Music Festival", "Music Mania Org."): "hasOrganizer",
    ("AI Conference", "2024-12-01 to 2024-12-03"): "hasTime",
    ("Blockchain Summit", "2025-01-10 to 2025-01-12"): "hasTime",
    ("Music Festival", "2025-02-15 to 2025-02-16"): "hasTime"
}

# izimi tekrar yap
plt.figure(figsize=(16, 14))
nx.draw(
    G_complex, complex_nodes, with_labels=True,
    labels={node: node for node in G_complex.nodes()},  # Node labels
    node_size=3500, node_color='lightblue',
    font_size=8, font_weight='bold', arrows=True, edge_color='purple'
)

# Kenar etiketlerini ekle
nx.draw_networkx_edge_labels(
    G_complex, complex_nodes, edge_labels=complex_edge_labels,
    font_color='red', font_size=8, label_pos=0.5
)

# Balk ekle
plt.title("Complex Ontology Visualization with Multiple Events", fontsize=16)
plt.show()
```
